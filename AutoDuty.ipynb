```python
import pandas as pd
import numpy as np
```

```python
l = [["P1","8:10-9:00"],
     ["P2","9:00-9:40"],
     ["P3","9:40-10:20"],
     ["SB","10:20-10:40"],
     ["P4","10:40-11:20"],
     ["P5","11:20-12:00"],
     ["P6","12:00-12:40"],
     ["LB","12:40-1:10"],
     ["P7","1:10-1:50"],
     ["P8","1:50-2:30"],
     ["P9","2:30-3:00"]]

c=['Name', 'P1','P2', 'P3', 'SB', 'P4', 'P5', 'P6', 'LB','P7', 'P8', 'P9']
final=[]
subjects=['Eng','Hindi','Fre-Sans','Kannada','Math','Phy','Chem','Bio','SS','Comp','VPA-LS','HPE']

for i in range(12):
    df = pd.DataFrame(pd.read_excel("StaffNew.xlsx",sheet_name=["Sheet"+str(i+1)])["Sheet"+str(i+1)])
    df.columns = c
    df = df[df.Name.notnull()]
    df=df.astype(str,errors='raise')
    subject=[]
    for j in range(len(df.index)):
        subject.append(subjects[i])
    df['subject']=subject
    final.append(df)

data = pd.concat(final)
```
